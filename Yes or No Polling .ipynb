```python
print('Example 1')

print('\nWelcome to the Yes or No Issue Polling App')

issue = input('\nWhat is the yes or no issue you will be voting on today: ')
voters = int(input('What is the number of voters you will allow on the issue: '))
password = input('Enter a password for polling results: ')

voting = {'Issue' : issue, 'Voters' : voters, 'Password': password}
votes = {}


for i in range(voting['Voters']):
    name = input('\nEnter your full name: ')
    
    if name not in votes.keys():
        print(f'Here is our issue: {issue}')
        answer = input('What do you think...yes or no: ').lower()
        if answer == 'yes' or answer == 'no':
            votes[name] = answer
            print(f'\nThankyou {name}! Your vote of {answer} has been recorded')
        else:
            print('That is not a yes or no answer, but okay...')
            votes[name] = answer
            print(f'\nThankyou {name}! Your vote of {answer} has been recorded')
    else:
        print(f'Sorry, it seems that someone with that name has already voted.')

print(f'\nThe following {len(votes)} people voted')


yes = 0
no = 0
for key,value in votes.items():
    if value == 'yes':
        yes += 1
    elif value =='no':
        no += 1
    print(key)

print(f'\nOn the following issue:{issue}:')
if yes > no:
    print(f'Yes wins! {yes} votes to {no}')
elif no > yes:
    print(f'No wins! {no} votes to {yes}')
else:
    print(f'It was a tie! {yes} votes to {no}.')

admin = input('\nTo see the voting results enter the admin password: ')
if admin == password:
    for key,value in votes.items():
        print(f'Voter : {key} \tVote : {value}')
else:
    print('Sorry, that is not the correct password. Goodbye...')
    
print('\nThank you for using the Yes or No Issue Polling App')
```
