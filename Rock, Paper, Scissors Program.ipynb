```python
import random

print('Welcome to a game of Rock, Paper, Scissors')

rounds = int(input('\nHow many rounds would you like to play: '))

player = 0
computer = 0

game = ['rock','paper','scissors']
random_element = random.choice(game)

for i in range(rounds):
    print(f'\nRound {i+1}')
    print(f'Player: {player} \t Computer: {computer}')
    print('Time to pick...rock, paper, scissors')
    player_choice = input('Player: ').lower()
    print(f'Computer: {random_element}')
    
    if random_element == 'rock':
        if player_choice == 'paper':
            print('paper covers rock')
            print(f'You win round {i+1}')
            player += 1
        elif player_choice == 'scissors':
            print('rock breaks scissors')
            print('Computer gets the point')
            computer += 1
        elif player_choice == 'rock':
            print('Its a tie, how boring')
        else:
            print('This is not a valid game option')
            print('Computer gets the point')
            computer += 1
    elif random_element == 'scissors':
        if player_choice == 'rock':
            print('Rock breaks scissors')
            print(f'You win round {i+1}')
            player += 1
        elif player_choice == 'paper':
            print('scissors cut paper')
            print('Computer gets the point')
            computer += 1
        elif player_choice == 'scissors':
            print('Its a tie, how boring')
        else:
            print('This is not a valid game option')
            print('Computer gets the point')
            computer += 1
    elif random_element == 'paper':
        if player_choice == 'rock':
            print('Paper covers rock')
            print('Computer gets the point')
            computer += 1
        elif player_choice =='scissors':
            print('scissors cut paper')
            print(f'You win round {i+1}')
            player += 1
        elif player_choice == 'paper':
            print('Its a tie, how boring')
        else:
            print('This is not a valid game option')
            print('Computer gets the point')
            computer += 1
            
print('\nFinal Game Results')
print(f'\tRounds Played: {rounds}')
print(f'\tPlayer Score: {player}')
print(f'\tComputer Score: {computer}')

winner = max(player,computer)
if player > computer:
    print(f'\tWinner: PLAYER!!')
elif player < computer:
    print(f'\tWinner: COMPUTER!!')
else:
    print(f'ITS A TIE!!')
```
