```python
import random

print('Welcome to the Coin Toss App')
print('\nI will flip a coin a set number of times.')
flips = int(input('How many times would you like me to flip the coin: '))

answer = input('Would you like to see the result of each flip (y/n): ').lower()

heads = 0
tails = 0

if answer == 'n':
    print('\nFlipping!')
    
    for i in range(1,flips+1):
        random_integer = random.randint(0,1)
        if random_integer == 0:
            tails += 1
        else:
            heads += 1
        if heads == tails:
            print(f'At {i} flips, the number of heads and tails were equal at {tails} each')
            
    print(f'\nResult of Flipping a coin {flips} times ')
    print(f'Side\t\tCount\t\tPercentage')
    print(f'Heads\t\t{heads}/{flips}\t\t{round((heads/flips)*100,2)}%')
    print(f'Tails\t\t{tails}/{flips}\t\t{round((tails/flips)*100,2)}%')
    
elif answer == 'yes' or answer == 'y' or answer == 'yeah':
    print('\nFlipping!')
    
    for i in range(1,flips+1):
        random_integer = random.randint(0,1)
        if random_integer == 0:
            print('TAILS')
            tails += 1
        else:
            print('HEADS')
            heads += 1
        if heads == tails:
            print(f'At {i} flips, the number of heads and tails were equal at {tails} each')
            
    print(f'\nResult of Flipping a coin {flips} times ')
    print(f'Side\t\tCount\t\tPercentage')
    print(f'Heads\t\t{heads}/{flips}\t\t{round((heads/flips)*100,2)}%')
    print(f'Tails\t\t{tails}/{flips}\t\t{round((tails/flips)*100,2)}%')
```
