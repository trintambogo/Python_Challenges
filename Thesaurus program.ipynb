```python
import random

print('Example 1')
print('Welcome to the Thesaurus App!')

thesaurus = {'hot':['balmy','summery','tropical','boiling','scorching'],
            "cold":['chilly', 'cool', 'freezing', 'frigid', 'polar'],
            "happy":['content', 'cheery', 'merry', 'jovial', 'jocular'],
            "sad":['unhappy', 'downcast', 'miserable', 'glum', 'melancholy']
            }

print('\nChoose a word from the thesaurus and I will give you a synonym.')
print('\nHere are the words in the thesaurus:')
for i in thesaurus:
    print(f'\t-{i}')
    
word = input('What word would you like a synonym for: ').lower()

if word not in thesaurus.keys():
    print("\nI'm sorry, that word is not currently in the thesaurus")
       
for key,value in thesaurus.items():
    if word == key:
        synonym = random.choice(value)
        print(f'The synonym for {key} is {synonym}')
        
answer = input('Would you like to see the whole thesaurus (yes/no): ').lower()
if answer == 'yes':
    for key,value in thesaurus.items():
        print(f'\n{key.title()} synonyms are:')
        for i in value:
            print(f'\t-{i}')
elif answer == 'no':
    print('\nI hope you enjoyed the program. Thank you!')


```
