```python
print('welcome to the Grade Point Average Calculator\n')
name = input('What is your name: ').title()
num_grades = int(input('How many grades would you like to enter:'))

grades = []

for i in range(num_grades):
    grade = float(input('Enter Grade: '))
    grades.append(grade)
    
print('\nGrades from highest to lowest: ')
for i in sorted(grades,reverse=True):
    print(f'\t{i}')
    
print(f'{name} Grade Total Summary')
print(f'\tTotal number of Grades; {num_grades}')
print(f'\tHighest grade: {max(grades)}')
print(f'\tLowest grade: {min(grades)}')
average = sum(grades)/len(grades)
print(f'\tAverage grade: {average}')

desired_average  = int(input('What is your desired average: '))
new_average_grade =  (desired_average*(len(grades)+1))-sum(grades)
print(f'\n GoodLuck{name}!')
print(f'You will need to get a {new_average_grade} on your next assignment to earn a {desired_average} average')

copy_grades = grades.copy()

print('Lets see what your average could have been if you did better/worse on an assignment.')
change_grade = float(input('What grade would you like to change: '))
new_grade = float(input(f'What grade would you like to change {change_grade} to : '))

copy_grades[copy_grades.index(change_grade)]=new_grade

print('\nGrades from highest to lowest: ')
for i in sorted(copy_grades,reverse=True):
    print(f'\t{i}')
    
print(f'{name} Grade Total Summary')
print(f'\tTotal number of Grades; {num_grades}')
print(f'\tHighest grade: {max(copy_grades)}')
print(f'\tLowest grade: {min(copy_grades)}')
new_average = sum(copy_grades)/len(copy_grades)
print(f'\tAverage grade: {new_average}')

print(f'Your new average would be {new_average} comparedto your real average {average}!')
print(f'That is a change of {round(new_average-average,1)} points!')

print(f'\nToo bad your original grades are still the same!')
print(grades)
print('You should go ask for extra credit!')
```
